```python
import mysql.connector
import pandas as pd

# Koneksi ke database MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="regis-website-ksl"
)

# Query SQL
query = """
SELECT
    users.name AS nama,
    pendaftarans.divisi_id AS divisi,
    CASE
        WHEN
            pendaftarans.user_id IS NOT NULL AND
            pendaftarans.divisi_id IS NOT NULL AND
            pendaftarans.prodi_id IS NOT NULL AND
            pendaftarans.nim IS NOT NULL AND
            pendaftarans.jk IS NOT NULL AND
            pendaftarans.tempat_lahir IS NOT NULL AND
            pendaftarans.tanggal_lahir IS NOT NULL AND
            pendaftarans.no_hp IS NOT NULL AND
            pendaftarans.foto IS NOT NULL AND
            pendaftarans.ktm IS NOT NULL AND
            pendaftarans.cv IS NOT NULL
        THEN 'True'
        ELSE 'False'
    END AS lengkap
FROM
    users
JOIN
    pendaftarans ON users.id = pendaftarans.user_id;
"""

# Membaca hasil query ke dalam DataFrame Pandas
df = pd.read_sql(query, conn)

# Menyimpan DataFrame ke file Excel
df.to_excel('hasil_query.xlsx', index=False)

# Menutup koneksi
conn.close()
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Buat koneksi ke MySQL menggunakan SQLAlchemy
engine = create_engine('mysql+mysqlconnector://root:@localhost/regis-website-ksl')

# Query SQL yang diperbarui
query = """
SELECT
    users.name AS nama,
    users.email AS email,
    pendaftarans.nim AS nim,
    pendaftarans.prodi_id AS prodi,
    pendaftarans.no_hp AS no_hp,
    pendaftarans.divisi_id AS divisi,
    CASE
        WHEN
            pendaftarans.user_id IS NOT NULL AND
            pendaftarans.divisi_id IS NOT NULL AND
            pendaftarans.prodi_id IS NOT NULL AND
            pendaftarans.nim IS NOT NULL AND
            pendaftarans.jk IS NOT NULL AND
            pendaftarans.tempat_lahir IS NOT NULL AND
            pendaftarans.tanggal_lahir IS NOT NULL AND
            pendaftarans.no_hp IS NOT NULL AND
            pendaftarans.foto IS NOT NULL AND
            pendaftarans.ktm IS NOT NULL AND
            pendaftarans.cv IS NOT NULL
        THEN 'True'
        ELSE 'False'
    END AS lengkap
FROM
    users
JOIN
    pendaftarans ON users.id = pendaftarans.user_id;
"""

# Membaca hasil query ke dalam DataFrame Pandas
df = pd.read_sql(query, engine)

# Memindahkan kolom 'lengkap' ke akhir
columns = [col for col in df.columns if col != 'lengkap'] + ['lengkap']
df = df[columns]

# Menyimpan DataFrame ke file Excel
df.to_excel('hasil_query.xlsx', index=False)

# Menutup koneksi (opsional karena SQLAlchemy menangani ini secara otomatis)
engine.dispose()

print("Data berhasil diubah dan disimpan ke file 'hasil_query_diperbarui.xlsx'.")
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Buat koneksi ke MySQL menggunakan SQLAlchemy
engine = create_engine('mysql+mysqlconnector://root:@localhost/regis-website-ksl')

# Query SQL yang diperbarui
query = """
SELECT
    users.name AS nama,
    users.email AS email,
    pendaftarans.nim AS nim,
    pendaftarans.prodi_id AS prodi,
    pendaftarans.no_hp AS no_hp,
    pendaftarans.divisi_id AS divisi,
    CASE
        WHEN
            pendaftarans.user_id IS NOT NULL AND
            pendaftarans.divisi_id IS NOT NULL AND
            pendaftarans.prodi_id IS NOT NULL AND
            pendaftarans.nim IS NOT NULL AND
            pendaftarans.jk IS NOT NULL AND
            pendaftarans.tempat_lahir IS NOT NULL AND
            pendaftarans.tanggal_lahir IS NOT NULL AND
            pendaftarans.no_hp IS NOT NULL AND
            pendaftarans.foto IS NOT NULL AND
            pendaftarans.ktm IS NOT NULL AND
            pendaftarans.cv IS NOT NULL
        THEN 'True'
        ELSE 'False'
    END AS lengkap
FROM
    users
JOIN
    pendaftarans ON users.id = pendaftarans.user_id;
"""

# Membaca hasil query ke dalam DataFrame Pandas
df = pd.read_sql(query, engine)

# Mengganti angka divisi
df['divisi'] = df['divisi'].replace({
    1: 'Programming',
    2: 'Multimedia',
    3: 'Networking'
})

# Mengganti angka prodi_id dengan nama program studi
df['prodi'] = df['prodi'].replace({
    1: 'Teknik Sipil',
    2: 'Pendidikan Vokasional Konstruksi Bangunan',
    3: 'Teknik Elektro',
    4: 'Teknik Komputer',
    5: 'Pendidikan Teknologi Informasi',
    6: 'Sistem Informasi',
    7: 'Pendidikan Seni Rupa',
    8: 'Perencanaan Wilayah dan Kota',
    9: 'Teknik Arsitektur',
    10: 'Teknik Industri',
    11: 'Pendidikan Teknik Mesin'
})

# Memindahkan kolom 'lengkap' ke akhir
columns = [col for col in df.columns if col != 'lengkap'] + ['lengkap']
df = df[columns]

# Menyimpan DataFrame ke file Excel
df.to_excel('hasil_query_diperbarui(1).xlsx', index=False)

# Menutup koneksi (opsional karena SQLAlchemy menangani ini secara otomatis)
engine.dispose()

print("Data berhasil diubah dan disimpan ke file 'hasil_query_diperbarui.xlsx'.")
```
